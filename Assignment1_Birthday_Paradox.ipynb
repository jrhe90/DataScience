```python
import random #import the Random Module
```

```python
def createBirthday(numiteration,numstudents): # To create a list of students' birthdays
    birthday =[[] for i in range(numiteration)] # the number of items in the list is based on the number of iterations
    for i in range(numiteration): 
        for x in range(numstudents): 
            ran = random.randint(1,365) # Generate the birthday beased on the random date from 1 to 365
            birthday[i].append(ran) # Add the random dates to an item [i] of the list (an item in the list is also a list)
    return birthday # To return a list of student birthdays, it's a list-of-lists
```
